```python
#Plotting training datas

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from statistics import mean


train_data_class1=np.array([[2,2],[3,1],[3,3],[-1,-3],[4,2],[-2,-2]])
train_data_class2=np.array([[-4,3],[2,6],[0,0],[-2,2],[-1,-1],[-4,2]])

df1 = pd.DataFrame(train_data_class1)
df2 = pd.DataFrame(train_data_class2)
#print(df1)
#print(df2)
```

```python
fig = plt.figure()
fig.set_size_inches(12, 6, forward=True)
axis = plt.subplot(111)

marker_size = 50

plt.scatter(df1[0],df1[1],color='red',marker='v',label="Train class1",s=marker_size)
plt.scatter(df2[0],df2[1],color='green',marker='*',label="Train class2",s=marker_size)


#Calculating mean values of both classes

Avg_class1=(mean(df1[0]),mean(df1[1]))
Avg_class2=(mean(df2[0]),mean(df2[1]))

#print(Avg_class1)
#print(Avg_class2)



plt.scatter(mean(df1[0]),mean(df1[1]),color='red',marker='o',label="Mean class1",s=marker_size)
plt.scatter(mean(df2[0]),mean(df2[1]),color='green',marker='s',label="Mean class2",s=marker_size)

test_data=np.array([[-1,-5,1],[3,2,1],[-2,1,2],[8,2,1],[6,-1,1],[0,2,1],[-3,0,2]])


c_df = pd.DataFrame(test_data)

#print(c_df)

test_class=c_df[2].tolist()
c_df.drop([2],axis=1,inplace =True)

pred_class1=pd.DataFrame()
pred_class2=pd.DataFrame()

pred_class=[]

for i in range(len(c_df)):
  g1=np.dot(c_df[i:i+1],Avg_class1)-0.5*np.dot(Avg_class1,Avg_class1)
  g2=np.dot(c_df[i:i+1],Avg_class2)-0.5*np.dot(Avg_class2,Avg_class2)
  if g1>g2:
    pred_class1=pred_class1.append(c_df[i:i+1])
    pred_class.append(1)
  else:
    pred_class2=pred_class2.append(c_df[i:i+1])
    pred_class.append(2)
  
#Ploting train data values
plt.scatter(pred_class1[0],pred_class1[1],color='red',marker='+',label="Test class1",s=marker_size)
plt.scatter(pred_class2[0],pred_class2[1],color='green',marker='x',label="Test class2",s=marker_size)


axis.legend()
plt.show()

```

```python
#Finding Accuracy

point=0
for i in range(len(pred_class)):
  if pred_class[i]==test_class[i]:
    point=point+1

accuracy=point/len(test_class)*100
print(accuracy)
```

```python
#Decision Boundary Between two classes

t = [Avg_class1[0]-Avg_class2[0],Avg_class1[1]-Avg_class2[1]]
x = np.linspace(min(c_df[0]), max(c_df[0]), 30)

y =  - ( t[0]*x + 0.5*(np.dot(Avg_class2 , Avg_class2) - np.dot(Avg_class1 , Avg_class1)) ) / t[1]
```

```python
fig = plt.figure()
fig.set_size_inches(14, 8, forward=True)
axis = plt.subplot(111)
marker_size = 50
plt.plot(x, y, color="red", linestyle='--', label = "Decision Boundary");
plt.scatter(df1[0],df1[1],color='red',marker='v',label="Train class1",s=marker_size)
plt.scatter(df2[0],df2[1],color='green',marker='*',label="Train class2",s=marker_size)
plt.scatter(mean(df1[0]),mean(df1[1]),color='red',marker='o',label="Mean class1",s=marker_size)
plt.scatter(mean(df2[0]),mean(df2[1]),color='green',marker='s',label="Mean class2",s=marker_size)
plt.scatter(pred_class1[0],pred_class1[1],color='red',marker='+',label="Test class1",s=marker_size)
plt.scatter(pred_class2[0],pred_class2[1],color='green',marker='x',label="Test class2",s=marker_size)

axis.legend()
plt.show()

```
